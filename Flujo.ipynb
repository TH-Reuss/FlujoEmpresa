```python
import pandas as pd
#import matplotlib.pyplot as plt
#%matplotlib inline

pd.options.display.max_rows = 300
```

```python
#Leemos el archivo
df_base = pd.read_csv('datos.csv')
#Eliminamos las filas con campos null
# df_base = df_base.dropna()
#Formateamos la fecha a pandas-date
df_base['Fecha'] = pd.to_datetime(df_base['Fecha'], format="%d/%m/%Y", errors='coerce')
df_base = df_base[['Fecha', 'Ingreso/Egreso', 'Monto']]
df_base = df_base.groupby([df_base['Fecha'].dt.to_period("M"), df_base['Ingreso/Egreso']]).agg('sum')
df_base.head(200)
```

```python
df_pivoted = df_base.pivot_table(index='Fecha', columns='Ingreso/Egreso', values='Monto', aggfunc='sum')
df_pivoted = df_pivoted.reset_index().rename_axis(None, axis=1)
df_pivoted = df_pivoted.set_index('Fecha')
df_pivoted.fillna(0, inplace=True)
#df_pivoted.head(200)
#df_pivoted.plot(kind = 'bar');
```

```python
df_pivoted['Utilidad'] = df_pivoted['Ingreso'] - df_pivoted['Egreso']
df_pivoted['Utilidad Acumulada'] = df_pivoted['Utilidad'].cumsum()
df_pivoted = df_pivoted[['Utilidad Acumulada']]
df_pivoted.head(20)

#df_pivoted.plot();
```
